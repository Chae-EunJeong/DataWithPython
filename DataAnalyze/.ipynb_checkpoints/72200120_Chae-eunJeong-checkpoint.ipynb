```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ten_year_married = pd.read_csv('married_in_Seoul.txt', delimiter='\t',dtype='int')
foreign_married = pd.read_csv('married_foreign_Seoul.txt',delimiter='\t',dtype='int')
ten_year_married.head()
del ten_year_married["조혼인율"]
ten_year_married.head()
```

```python
del foreign_married["한국인 남편+외국인 아내"]
del foreign_married["한국인 아내+외국인 남편"]

foreign_married.head()
```

```python
ten_year_married = ten_year_married.pivot("자치구","기간","혼인건수")
ten_year_married.head(10)
```

```python
foreign_married = foreign_married.pivot("자치구","기간","외국인과의 혼인 계")
foreign_married.head(14)
```

```python
foreign_married = foreign_married.astype({'2010':int, '2011':int, '2012':int, '2013':int, '2014':int, '2015':int, '2016':int, '2017':int, '2018':int, '2019':int})
foreign_married.dtypes
```

```python
perc = int(foreign_married.div((int)ten_year_married))
perc
```
